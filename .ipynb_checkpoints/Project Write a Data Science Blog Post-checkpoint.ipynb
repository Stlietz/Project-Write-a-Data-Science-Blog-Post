##  `2.2` PCA - Review scores

Since most of the review scores are correlated with each other, and as well primarily uncorrelated or only to a minor extent with price, I want to summarize the information in principle components, to be able to reduce the feature space if needed resp. see if they can contribute to predict price taken together.

```python
PCA_vars = review_cols
g = sns.PairGrid(data = df_prices, vars = PCA_vars)
g.map_diag(plt.hist)
g.map_lower(plt.scatter, alpha = 0.6)
g.map_upper(sns.regplot);
# all features show a more or less visible linear relationship
```

```python
scaler = StandardScaler()
df_PCA = df_prices[review_cols]
df_PCA.head()
```

```python
# Scaling the data for pca
scaler.fit(df_PCA)
scaled_data = scaler.transform(df_PCA)
print(scaled_data.shape)
scaled_data
```

```python
# Apply PCA to the data.
def implement_pca(n_components, data):
    '''
    Transforms data using PCA to create n_components, and provides back the results of the transformation
    
    INPUT: n_components - int - the number of principal components to create
           data - the data to be transformed
           
    OUTPUT: pca - the pca object created after fitting the data
            X_pca - the transformed X matrix with the new number of components
    '''
    
    pca = PCA(n_components, random_state=42)
    X_pca = pca.fit_transform(data)
    
    return pca, X_pca
```

```python
pca, data_transformed = implement_pca(scaled_data.shape[1], scaled_data)
print(data_transformed.shape[0] == df_prices.shape[0])
data_transformed
```

```python
print(pca.components_.shape)
pca.components_
```

```python
# Investigate the variance accounted for by each principal component.

num_components = len(pca.explained_variance_ratio_)
print(num_components)

ind = np.arange(num_components)
vals = pca.explained_variance_ratio_

cumvals = np.cumsum(vals)

bins = np.arange(0, num_components)
x_label = bins+ 1

plt.figure(figsize=(8,6))

plt.bar(range(num_components), vals, alpha = 0.5, align="center", label="individual explained variance")
plt.step(range(num_components), cumvals, where="mid", label="cumulative explained variance")
plt.ylabel("Explained Variance Ratio")
plt.xlabel("Principal Component")
plt.xticks(ticks=bins, labels=x_label)
plt.legend(loc = "best");

print("explained variance :", vals)
print("cumulated explained variance :", cumvals)
```

The first component accounts for approx. 46% of the variability. The first two (resp. three) about 61% (71%). Let's look at the correlation between the PC's and price

```python
def train_test_split_data(df = df_prices, exclude_x = ["id"], target=["price"], test_s = 0.3, random=42):
    """split data into train and test"""
    X = df.drop(exclude_x + target, axis=1)
    y = df[target]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = test_s, random_state=random)
    return X_train, X_test, y_train, y_test
```

```python
def get_hyperparam_grid(i = 0.001, limit = 10, step=10):
    """construct alpha parameter to check in lasso"""
    alphas = [i]
    
    while i != limit:
        i = i*step
        alphas.append(i)

    parameters = {"alpha": alphas}

    return parameters

alphas = get_hyperparam_grid()
alphas
```

```python
def print_best_alpha_cv(rng=1337, params=alphas, cv_s = 3, test_s=0.3):
    """
    perform Cross Validation using a Lasso Model with different alpha values to find best model

    print the best alpha and the train and test errors
    """

    X_train, X_test, y_train, y_test = train_test_split_data(test_s=test_s, random=rng)
    
    
    lasso = Lasso(random_state=rng)
    lasso_regressor = GridSearchCV(lasso, alphas, cv = cv_s)

    lasso_regressor.fit(X_train, y_train)

    # best alpha
    print("Best Alpha: ", lasso_regressor.best_params_["alpha"], "\n")

    # get scores / R^2
    train_score = lasso_regressor.score(X_train,y_train)
    test_score = lasso_regressor.score(X_test,y_test)
    
    print("Training score: {:2f}%".format(train_score*100))
    print("Test score: {:2f}%".format(test_score*100))
```

```python
print_best_alpha_cv()
```

## `3.3` Resampling Lasso Regression - Significant coefficients and Drivers of Listing Prices

```python
df_prices.head()
```

```python
def bootstrap_lasso(df=df_prices, alpha=0.01, exclude_x = ["id"], target="price", boot_iterations=5000, max_iter = 100, set_warnings="ignore",
                    print_time=True):


    import warnings
    warnings.filterwarnings(set_warnings)

    # list to store model coefficients
    coef_data = []
    
    # exclude cols if specified
    df = df.drop(exclude_x, axis=1)

    # set up df
    ind_vars = df.drop(target, axis=1).columns
    
    #np.random.seed(seed)
    # seed sequence for resampling  
    seeds = [i for i in range(boot_iterations)]

    # Bernoulli sample n values from original data sample of n values. Re-sample = size of df
    n_samples = df.shape[0] 

    # for each bootstrap iteration
    for i in range(boot_iterations):

        # resample with replacement, i.e. bootstrap
        model_sample = df.sample(n_samples, replace=True, random_state=seeds[i])
        
        lasso = Lasso(alpha = alpha)

         # fit model on sample
        lasso.fit(model_sample.drop(target, axis=1), model_sample[target])
        coef_data.append(lasso.coef_)

    df_coef_boot = pd.DataFrame(coef_data)
    
    df_coef_boot.columns = ind_vars
    
    return df_coef_boot
```

```python
df_coef_boot = bootstrap_lasso(boot_iterations=5000)
```
